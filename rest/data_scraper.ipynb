```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time

# Set up Selenium WebDriver (Make sure to have ChromeDriver installed)
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run in headless mode (no GUI)
service = Service("/usr/bin/chromedriver")  # Update with your correct path
driver = webdriver.Chrome(service=service, options=options)

# Open the first page
driver.get("https://www.zakonyprolidi.cz/obor/trestni-pravo")

df = pd.DataFrame()
page_num = 1

while True:
    time.sleep(2)  # Wait for JavaScript to load content

    # Get page source and parse with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    table = soup.find_all('table')[4]  # Adjust index if needed

    if not table:
        break

    world_titles = table.find_all('th')
    if page_num == 1:
        world_table_titles = [title.text.strip() for title in world_titles] + ["Full Text"]
        df = pd.DataFrame(columns=world_table_titles)

    rows = table.find_all('tr')[1:]

    for row in rows:
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]

        # Extract law link
        link_element = row.find('a')
        full_text = ""
        if link_element:
            predpis_url = "https://www.zakonyprolidi.cz" + link_element['href']
            driver.get(predpis_url)
            time.sleep(2)  # Wait for content

            predpis_soup = BeautifulSoup(driver.page_source, 'html.parser')
            law_text_element = predpis_soup.find("div", class_="law-content")  # Adjust class
            if law_text_element:
                full_text = law_text_element.get_text(separator=" ").strip()

            driver.back()  # Return to main page

        individual_row_data.append(full_text)
        df.loc[len(df)] = individual_row_data

driver.quit()
print(df)
```
